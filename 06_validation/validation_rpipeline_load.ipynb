```r
# make function for yaml parsing

task_location <- function(pipeline_yaml, search_string) {
    task_location = 0
    for (n in seq(length(pipeline_yaml$tasks))) {
        if (pipeline_yaml$tasks[[n]]$name == search_string) {
            task_location = n
    }}
    return(task_location)
}
```

```r
task_location(pipeline_yaml, "Compute DASH")
```

```r
for (studyType in studyList) {
    print(pipeline_yaml$tasks[[task_location(pipeline_yaml, "Compute DASH")]]$product[[studyType]]) 
    }
```
